```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 160
import pickle
file = open("../../rust_sonic_surface/field.pickle", "rb")
field = pickle.load(file)
```

```python
F = np.array(field)
F = F[:, :, :, 0] + 1j * F[:, :, :, 1]
norm = np.abs(F)/np.max(np.abs(F))
phase = np.angle(F)
```

```python
cut = 20
plt.pcolormesh(np.linspace(0, 0.16, N), np.linspace(0, 0.1, N)[:-cut], norm[:, 50, :].swapaxes(0, 1)[:-cut, :])
plt.scatter(0.075, 0.05, 50, c='b', marker='o')
plt.colorbar()
plt.xlabel("x (m)")
plt.ylabel("z (m)")
plt.title(f"Normalized pressure distribution for a trap")
```

```python
cut = 40
plt.pcolormesh(np.linspace(0, 0.1, N), np.linspace(0, 0.1, N)[:-cut], phase[50, :, :].swapaxes(0, 1)[:-cut, :])
plt.scatter(0.0505, 0.05, 50, c='b', marker='o')
plt.colorbar()
plt.xlabel("x (m)")
plt.ylabel("z (m)")
plt.title(f"Normalized pressure distribution for a trap")
```
